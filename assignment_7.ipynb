### 1.1 What is the expected value of a single die roll? 

Recall that the expected value of a random variable $X$ is
$$
\mathbb{E}[X] = \begin{cases} \int_{x \in \text{supp}(X)} x \times f_X(x)dx, & \text{$X$ continuous}\\
\sum_{x \in \text{supp}(X)} x \times m_X(x), & \text{$X$ discrete}.
\end{cases}
$$

$$

### 1.3 What is the expected winnings of any gamble in European roulette?

In European roulette, we have 37 slots for the ball to land in (1-36 and a single 0). There is an equal probability of landing in each slot.

So, if we pay one dollar to bet on K slots, and we have probability of $\frac{K}{37}$, if we win we get $\frac{36}{K}$ dollars and if we lose, which has a probability of $\frac{(37-K)}{37}$, we get $0$ dollars.

So, the expected winnings for any bet is 

$$\mathbb{E}[\text{Winnings}]=(\text{win probability})(\text{win payoff})+(\text{lose probability})(\text{lose payoff})-(\text{initial bet})$$

We plug in our probabilities and payoffs to find
$$\mathbb{E}[\text{Winnings}]=\left(\frac{K}{37}\right)\left(\frac{36}{K}\right)+\left(\frac{(37-K)}{37}\right)(0) - 1$$

Simplify
$$\mathbb{E}[\text{Winnings}]=\frac{36}{37}-1=-0.027$$

So, any one-dollar roulette bet has an expected payoff of $-0.027$ dollars.

### 2.1 Compute the expected value for a uniform random variable

We know that the PDF for a uniform random variable is 

$$
f(x) =\frac{1}{b-a}, \quad a\le x \le b
$$

The support is $[a,b]$.

Recall that the expected value of a random variable $X$ is
$$
\mathbb{E}[X] = \begin{cases} \int_{x \in \text{supp}(X)} x \times f_X(x)dx, & \text{$X$ continuous}\\
\sum_{x \in \text{supp}(X)} x \times m_X(x), & \text{$X$ discrete}.
\end{cases}
$$

If we consider a continous variable, we will use an integral. And since our probability is 0 everywhere except between $a$ and $b$, we can bound our integral.

So, we have
$$
\mathbb{E}[X] = \int_a^b x \times \left(\frac{1}{b-a}\right)dx
$$

We pull out the constant $\frac{1}{b-a}$
$$
\mathbb{E}[X] = \left(\frac{1}{b-a}\right) \int_a^b x dx
$$

We integrate
$$
\mathbb{E}[X] = \left(\frac{1}{b-a}\right) \left[\frac{x^2}{2}\right]_a^b
$$

And evaluate at the bounds
$$
\mathbb{E}[X] = \left(\frac{1}{b-a}\right) \left(\frac{b^2}{2}-\frac{a^2}{2}\right)=\left(\frac{1}{b-a}\right) \left(\frac{b^2-a^2}{2}\right) = \frac{b^2-a^2}{2(b-a)}
$$

Recall difference of squares
$$
\mathbb{E}[X] = \frac{(b-a)(b+a)}{2(b-a)}
$$

Cancel out $(b-a)$ terms
$$
\mathbb{E}[X] = \frac{b+a}{2}
$$

### 2.2 Show that $\mathbb{E}[a+bX] = a + b\mathbb{E}[X]$

We need to prove the linearity of the expectation. We want to show that if we scale a variable $(bX)$ and shift it $(a)$, the average scales and shifts exactly the same way.

$$\mathbb{E}[a+bX] =\int_{-\infty}^\infty (a+bx)f(x)dx$$

Break the integral into two parts (integrals are linear operators so we can do this)
$$\mathbb{E}[a+bX] =\int_{-\infty}^\infty af(x)dx+ \int_{-\infty}^\infty bxf(x)dx$$

Pull out the constants $a$ and $b$
$$\mathbb{E}[a+bX] =a\int_{-\infty}^\infty f(x)dx+ b\int_{-\infty}^\infty xf(x)dx$$

Notice that 
$\int_{-\infty}^\infty f(x)dx$ is the total area under the curve and thus equals $1$ (because its probability) and, by definition, that $\int_{-\infty}^\infty xf(x)dx$ is $\mathbb{E}[X]$.

So, we have
$$\mathbb{E}[a+bX] =a(1)+b\mathbb{E}[X]=a+b\mathbb{E}[X]$$

### 8.2 - Write out the likelihood

Since our patients are independent, we multiply the probabilities together.

We know that the likelihood function is the joint density of the error terms, by independence. (More simply, its how likely is it that we would see a specific combination of data points, given a specific parameter $p$.)

The formula for likelihood is 
$$L(p) = \prod_{i=1}^n f_{X_i}(x_i; p)$$

$$L(p) = \prod_{i=1}^n p^{y_i}(1-p)^{1-y_i}$$

### 8.3 Maximize the likelihood with respect to $p$. What is the MLE, $\hat{p}$?

Maximizing is an optimization problem, so we will take the derivative, set it to 0, and solve for $\hat{p}$. This will be easier if we maximize the log likelihood instead. Since log is a strictly increasing function, the peak for the log likelihood will be at the same place as the peak for the regular likelihood.

To maximize, we first take logs. Which allows us to split our multiplication up into addition.
$$
\ell(p) = \sum_{i=1}^n y_i \log(p) + \left(1-\sum_{i=1}y_i\right) \log(1-p)
$$

Now let's take the derivative
$$
\ell'(p) = \frac{1}{p}\sum_{i=1}^n y_i - \frac{1}{1-p}\left(1-\sum_{i=1}^ny_i\right) 
$$

And set it to zero and solve for $\hat{p}$
$$
\frac{1}{\hat{p}} \sum_{i=1}^n y_i - \frac{1}{1-\hat{p}}\left(1-\sum_{i=1}^ny_i\right)  = 0
$$

This simplifies to
$$
\frac{\sum_{i=1}^n y_i}{\hat{p}} - \frac{\left(1-\sum_{i=1}^ny_i\right)}{1-\hat{p}}  = 0
$$

Moving the right fraction to the right hand side, we have
$$
\frac{\sum_{i=1}^n y_i}{\hat{p}} = \frac{\left(1-\sum_{i=1}^ny_i\right)}{1-\hat{p}}
$$

Multiplying both sides by $\hat{p}(1-\hat{p})$ gives
$$
(1-\hat{p})\sum_{i=1}^n y_i = \hat{p}\left(1-\sum_{i=1}^ny_i\right)
$$

This simplifies to
$$
\sum_{i=1}^n y_i-\hat{p}\sum_{i=1}^n y_i = \hat{p}-\hat{p}\sum_{i=1}^ny_i
$$

Add $\hat{p}\sum_{i=1}^ny_i$ to both sides to get our MLE
$$
\hat{p}=\sum_{i=1}^n y_i
$$

### 8.4 - Bootstrap the sampling density/distribution of $\hat{p}$

```python
# bootstrap the sampling distribution
# convert data from string to numeric
y = pd.to_numeric( y.replace({'NO':'0','YES':'1'}) , errors='coerce')

# return the original MLE
print(f'Original MLE: ', y.mean())

# set the number of simulations
n_sim = 5000
n_size = len(df) # size of original dataset

# create an empty list for the findings
boot_dist_p = []

# loop through the number of simulations
for i in range(n_sim):
    # sample with replacement from the original dataset
    sample = np.random.choice(y, size=n_size, replace=True)

    # get and store the MLE of each sample
    boot_dist_p.append(sample.mean())
    
# convert to numpy array
boot_dist_p = np.array(boot_dist_p)

# visualize results
plt.figure(figsize=(10, 6))
sns.histplot(boot_dist_p, stat='density', kde=True)
plt.axvline(y.mean(), color='red', linestyle='--', label=f'Original MLE: {y.mean():.3f}')
plt.title('Bootstrapped Sampling Distribution of p_hat')
plt.xlabel('Estimated Probability (p)')
plt.legend()
plt.show()
```
